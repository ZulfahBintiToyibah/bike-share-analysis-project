### Cleaning Data

#### Memperbaiki Tipe Data dan Values (Nilai)
Konversi **dteday ke datetime** - Kolom tanggal dikonversi ke tipe datetime untuk memudahkan analisis waktu

```python
# Konversi Kolom 'dteday' Menjadi Format Tanggal
df_day['dteday'] = pd.to_datetime(df_day['dteday'])
df_day.info()
```

Memperbaiki values (nilai) dari kolom **weekday, mnth, yr, season**

```python
# Mengonversi nilai pada kolom 'weekday' menjadi nama hari yang sesuai
df_day['weekday'] = df_day['dteday'].dt.strftime('%A')

# Mengonversi nilai pada kolom 'mnth' menjadi nama bulan yang sesuai
df_day['mnth'] = df_day['dteday'].dt.strftime('%B')

# Memperbarui nilai pada kolom 'yr' menjadi tahun yang sesuai
df_day['yr'] = df_day['dteday'].dt.year

# Mengganti nilai pada kolom 'season' dengan nama musim yang relevan
df_day['season'].replace({1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}, inplace=True)
```

#### Normalisasi Kolom Numerik
Menggunakan MinMaxScaler untuk menormalkan **temp, atemp, hum, dan windspeed**.

```python
# Normalisasi Data Numerik pada Kolom 'temp', 'atemp', 'hum', dan 'windspeed'
scaler = MinMaxScaler()

for df in [df_day]:
    df[['temp', 'atemp', 'hum', 'windspeed']] = scaler.fit_transform(df[['temp', 'atemp', 'hum', 'windspeed']])

print("Data setelah normalisasi (menampilkan beberapa baris teratas):")
print(df_day[['temp', 'atemp', 'hum', 'windspeed']].head())
```

#### Validasi Rentang Data
Memastikan **weathersit** berada dalam rentang yang diharapkan (1-4) sehingga data di luar rentang tersebut akan dihapus



```python
print("\nValidasi rentang data pada kolom kategori...")

# Validasi kolom 'weathersit' harus memiliki nilai antara 1 hingga 4
print("Cek 'weathersit' pada df_day:", df_day[~df_day['weathersit'].between(1, 4)])

# Menghapus data di luar kategori yang valid untuk 'weathersit' saja
df_day = df_day[df_day['weathersit'].between(1, 4)]

print("\nProses Data cleaning Telah selesai. Data siap untuk analisis lebih lanjut.")
```

```python
# Data df_day yang sudah melalui tahapan data cleaning
df_day.head()
```

Setelah dilakukan tahapan data cleaning pada **'df_day'**, dimana kita telah melakukan beberapa tahapan diantaranya memperbaiki tipe data pada kolom **'dteday'** menjadi format tanggal. Selanjutnya juga dilakukan konversi values (nilai) dari kolom **weekday, mnth, yr, season**. Dapat dilihat kolom weekday sudah berubah menjadi nama hari (sunday, monday, tuesday, wednesday, thursday, friday, dan saturday). Selanjutnya pada kolom mnth dan yr sudah sesuai dengan nama bulan dan tahun. Kolom season juga sudah berubah menjadi nama musim yang sesuai (summer, spring, fall, winter). Selain itu juga dilakukan normalisasi kolom pada **'temp', 'atemp', 'hum', dan 'windspeed'** serta melakukan validasi rentang data

### Pertanyaan 1: Bagaimana tren penggunaan sepeda dalam jangka Waktu setahun? mana yang lebih unggul tahun 2011 atau 2012?

```python
# Mengelompokkan data berdasarkan akhir setiap bulan ('ME') menggunakan tanggal pada kolom 'dteday'
df_year = df_day.resample(rule='ME', on='dteday').agg({
    "casual": "sum",
    "registered": "sum",
    "cnt": "sum"
})
df_year.head()
```

```python
# Mengubah indeks DataFrame 'df_year' menjadi format string bulan-tahun (contoh: 'Jan-22')
df_year.index = df_year.index.strftime('%b-%y')

# Mereset indeks pada DataFrame dan menyimpannya dalam variabel baru 'df_year_users'
df_year_users = df_year.reset_index()

# Mengubah nama kolom 'dteday' menjadi 'yearmonth'
df_year_users = df_year_users.rename(columns={'dteday': 'yearmonth'})
```

```python
# Menentukan ukuran gambar untuk plot
plt.figure(figsize=(12, 6))

# Membuat grafik garis dengan seaborn, x = 'yearmonth' dan y = 'cnt'
sns.lineplot(x="yearmonth", y="cnt", data=df_year_users, color='purple', marker='o', linestyle='--')

# Menambahkan label pada sumbu x dan y serta judul grafik
plt.xlabel("Bulan-Tahun")
plt.ylabel("Jumlah Penyewa Sepeda")
plt.title("Grafik Jumlah Penyewa Sepeda Tahun (2011-2012)")

# Memutar label pada sumbu x sebesar 45 derajat untuk keterbacaan
plt.xticks(rotation=45)

# Menyesuaikan tata letak agar elemen-elemen tidak terpotong
plt.tight_layout()

# Menampilkan grafik
plt.show()
```

**Insight:**

Berdasarkan hasil visualisasi menggunakan line plot, grafik diatas menunjukkan bahwa tren penggunaan sepeda awal tahun memiliki jumlah penyewaan yang tergolong masih rendah, kemudian terjadi peningkatan di pertengahan tahun dan mengalami penurunan di akhir tahun. Sehingga dapat disimpulkan bahwa **grafik tahun 2011 cenderung lebih rendah daripada tahun 2012**, tahun 2012 menjadi tahun yang unggul pada penyewaan sepeda.

### Pertanyaan 4: Hari apa penggunaan sepeda paling banyak dan paling sedikit?

```python
# Mengelompokkan data berdasarkan hari dalam seminggu ('weekday') dan menghitung jumlah penyewa sepeda
weekday_df = df_day.groupby('weekday')[['registered', 'casual','cnt']].sum().reset_index()
weekday_df.head(6)
```

```python
# Menyusun ulang dataFrame 'weekday_df' untuk membuat kolom 'status' yang menggabungkan
day = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_df_detail = pd.melt(weekday_df,
                            id_vars=['weekday'],
                            value_vars=['registered', 'casual'],
                            var_name='status',
                            value_name='count')
```

```python
# Visualisasi data
plt.figure(figsize=(9, 5))

sns.barplot(x='weekday', y='count', data=weekday_df_detail, order=day, hue='status', palette="magma")

plt.xlabel("Hari dalam Seminggu")
plt.ylabel("Jumlah Penyewaan Sepeda")
plt.title("Jumlah Penyewaan Sepeda Berdasarkan Hari dalam Seminggu")

plt.show()
```

**Insight:**

Berdasarkan hasil visualisasi bar plot, grafik diatas menunjukkan bahwa penyewa sepeda (baik registered dan casual) penggunaan paling banyak pada hari kerja (weekday) yaitu di hari **Jumat (friday)** dan penggunaan paling sedikit pada hari **minggu (sunday)**

## Analisis Lanjutan (Opsional)

Melakukan analisis menggunakan **Clustering** terhadap frekuensi penyewa sepeda per-hari

```python
# Batasan untuk rendah (0 - < 1000), sedang (antara 1000-5000), dan tinggi (> 5000)
batas = [0, 1000, 5000, float('inf')]
kategori = ['Rendah', 'Sedang', 'Tinggi']
```

```python
# Menambahkan kolom 'kategori' ke DataFrame df_day berdasarkan nilai 'cnt'
df_day['kategori'] = pd.cut(df_day['cnt'], bins=batas, labels=kategori, right=False)
```

```python
# Menghitung jumlah data untuk setiap kategori dan mereset indeksnya
jumlah_per_kategori = df_day['kategori'].value_counts().reset_index()

# Mengubah nama kolom pada DataFrame hasil
jumlah_per_kategori.columns = ['kategori', 'jumlah']

# Menampilkan hasil jumlah per kategori
print(jumlah_per_kategori)
```

```python
# Menentukan ukuran gambar
plt.figure(figsize=(7, 5))

# Membuat grafik batang horizontal dengan seaborn, x = jumlah, y = kategori
sns.barplot(x='jumlah', y='kategori', data=jumlah_per_kategori, palette='Set2')

# Menambahkan label pada sumbu x dan y, serta judul grafik
plt.xlabel("Jumlah Penyewa Sepeda")
plt.ylabel("Kategori")
plt.title("Jumlah Penyewa Sepeda Berdasarkan Kategori")

plt.tight_layout()

plt.show()
```


* Kategori **tinggi** memiliki 286 tupel/baris
* Kategori **sedang** memiliki 427 tupel/baris
*   Kategori **rendah** memiliki 18 tupel/baris
* Dapat diperoleh kesimpulan bahwa **kategori sedang** memiliki jumlah penyewa sepeda **paling banyak** tiap harinya dengan rentang 1000-5000
